```python
from dotenv import load_dotenv
import os

load_dotenv()

Goole_API_KEY=os.getenv('GOOGLE_API_KEY')

os.environ['GOOGLE_API_KEY']=Goole_API_KEY


from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="models/gemini-2.5-flash",
    temperature=0.7
)
name=llm.invoke('i want to opne a resturant for indian food.Suggest a fency name for this')
print(name.content)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template_name=PromptTemplate(
    input_variables=['cuisine'],
    template="i want to opne a resturant for {cuisine} food.Suggest a fency name for this"
    
)

prompt_template_name.format(cuisine="Italian")




from langchain_core.output_parsers import StrOutputParser

chain = prompt_template_name | llm | StrOutputParser()

result = chain.invoke({"cuisine": "Mexican"})
print(result)
```
